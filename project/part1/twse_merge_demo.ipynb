```python
#使用CSV
import pandas as pd

#讀取要合併的csv檔
csv1 = pd.read_csv('fp_demo.csv')
csv2 = pd.read_csv('fp_demo2.csv')

#以'日期'為index合併Dataframe
result = pd.merge(csv1, csv2, on='日期')
result.head()
```

```python
result.to_csv('fp_demo3.csv', sep=',', encoding='utf-8-sig', index=False)
```

```python
#使用SqLite
import pandas.io.sql as pd_sql
import sqlite3 as sql
import pandas as pd

#連線至sqlite檔案，若無該檔案sql，則會建立一個新的
conn = sql.connect("twse.db")


df1 = pd.read_sql_query("select * from demo1;", conn)
df2 = pd.read_sql_query("select * from demo2;", conn)

#以'日期'為index合併Dataframe
result = pd.merge(df1, df2, on='日期')
#刪除多餘欄位
result = result.drop(['index_x', 'index_y'], axis=1)
#將result寫入sql檔中的'demo3'表中，無該資料表則會自動建立
result.to_sql("demo3", conn, if_exists="replace")
```
